```python
def update_word(word: tuple[bytes], pair: tuple[bytes]) -> tuple[tuple[bytes], list[tuple[bytes]], list[tuple[bytes]]]:
    new_token = pair[0] + pair[1]
    new_word = []

    i = 0
    while i < len(word):
        if i < len(word) - 1 and (word[i], word[i + 1]) == pair:
            new_word.append(new_token)
            i += 2
        else:
            new_word.append(word[i])
            i += 1

    old_pairs = [(word[j], word[j + 1]) for j in range(len(word) - 1)]
    new_pairs = [(new_word[k], new_word[k + 1]) for k in range(len(new_word) - 1)]

    return tuple(new_word), new_pairs, old_pairs
```

```python
update_word(
    (b' ', b't', b'h', b'e', b'r', b'm', b'e', b'r', b'e', b't', b'e', b'r'), 
    (b'e', b'r'))
```
